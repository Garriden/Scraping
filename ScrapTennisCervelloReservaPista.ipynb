```python
def sendMail():
    # Define email sender and receiver
    email_sender = 'nbetsofficial@gmail.com'
    email_password = "Animalnumero!"
    #email_receiver = ['Garridocho@gmail.com', 'aliahsan365@gmail.com', 'batrolonews@gmail.com']
    email_receiver = ['Garridocho@gmail.com', 'batrolonews@gmail.com']
    
    # Set the subject and body of the email
    subject = 'Script Reserva tennis Dijous a les 20h'
    body = """
    
    URL = "https://reservadeportes.com/ClubTennisCervello.html" 
    
    
    Sóc l'script de'n Marc!
    
    Doncs sembla que ha anat bé.
    
    Tens reservada la pista 1 del proxim dijous a les 20h.
    
    
    
    """

    em = EmailMessage()
    em['From'] = email_sender
    em['To'] = ", ".join(email_receiver)
    em['Subject'] = subject
    em.set_content(body)

    # Add SSL (layer of security)
    context = ssl.create_default_context()

    # Log in and send the email
    with smtplib.SMTP_SSL('smtp.gmail.com', 465, context=context) as smtp:
        smtp.login(email_sender, email_password)
        smtp.sendmail(email_sender, email_receiver, em.as_string())
```

```python
def OpenWindow(URL):
    #options = webdriver.ChromeOptions()
    #driver = webdriver.Chrome(ChromeDriverManager().install(), chrome_options=options)
    
    driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
    #print (webdriver.__version__)

    return driver
```

```python
def ReservaPista(driver):

    driver.get(URL)

    time.sleep(5)
    
    ############
    ## LOG IN
    
    username = 'garridocho@gmail.com'
    password = "yuyuyuyu"
    
    usernameInput = driver.find_element("id",   "username")
    passwordInput = driver.find_element("id",   "password")
    submit        = driver.find_element("xpath", "//button[@class='mybuttonclass btn btn-submit']")    

    if usernameInput and passwordInput and submit:
        usernameInput.send_keys(username)
        passwordInput.send_keys(password)
        
        time.sleep(2)
        submit.click();
        time.sleep(1)
    else:
        print("Could NOT log in!")   
        
    ##########################
    ## Click into Reserving page
    
    reservaPageButton = driver.find_element("xpath", "//div[@class='mybuttonclass']")  
    if reservaPageButton:
        time.sleep(2)
        reservaPageButton.click();
        time.sleep(1)
    else:
        print("Could NOT enter the reserving page!") 
    
    dayButton = driver.find_element("xpath", "//a[contains(.,'Jueves')]")
    if dayButton:
        time.sleep(2)
        dayButton.click();
        time.sleep(1)
    else:
        print("Could NOT Click the day!") 
        
    hourButton = driver.find_element("xpath", "//a[contains(.,'20:00 > 21:00')]")
    if hourButton:
        time.sleep(2)
        hourButton.click();
        time.sleep(1)
    else:
        print("Could NOT Click the hour!") 
        
    
    ###############################
    ## MAIN PAGE 
    ## Reservar pista
        
    individualCombobox = Select(driver.find_element("xpath", "//select[contains(.,'Indi')]"))
    if individualCombobox:
        time.sleep(2)
        individualCombobox.select_by_index(1) # Individuales.
        time.sleep(2)
        
    else:
        print("Could NOT Click the individualCombobox!") 
        
    player = driver.find_element("xpath", "//option[contains(.,'GARRIDO GALBAS')]")
    if player:
        time.sleep(2)
        player.click()
        time.sleep(2)
        
    else:
        print("Could NOT Click the right player!") 
    
    terms = driver.find_elements("xpath", "//input[@type='checkbox']")
    if terms:
        time.sleep(1)
        terms[1].click() # Second checkbox.
        time.sleep(2)  
    else:
        print("Could NOT Click the terms check!") 

    
    accept = driver.find_element("xpath", "//input[@id='btnsubmitform']")
    if accept:
        time.sleep(1)
        accept.click()
        time.sleep(2)     
    else:
        print("Could NOT Click the accept button!") 
    
    print("SUCCESS!")   

    return True
```

```python
URL = "https://reservadeportes.com/ClubTennisCervello.html" 

driver = OpenWindow(URL)

reservaDone = ReservaPista(driver)

if(reservaDone):
    sendMail()
```
